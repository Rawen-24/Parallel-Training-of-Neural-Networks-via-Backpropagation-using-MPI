```python
import csv
import numpy as np

csv_file = "/home/rawen/2025-ba-jendoubi/.vscode/Dataset/mnist_test.csv"
output_file = "mnist_test_dataset.dat"
labels_file = "mnist_test_labels.dat"
pixels_file = "mnist_test_pixels.dat"
num_classes = 10  # MNIST digits 0-9

# Set to None to extract ALL samples, or set a number e.g. 1000, 20000
num_samples = 1000

all_labels = []
all_pixels = []

with open(csv_file, "r") as f_in, open(output_file, "w") as f_out:
    reader = csv.reader(f_in)
    next(reader)  # Skip header
    for i, row in enumerate(reader):
        if num_samples is not None and i >= num_samples:
            break

        label = int(row[0])
        pixels = np.array(row[1:], dtype=np.float32) / 255.0  # Normalize to [0,1]

        # One-hot encoding as integers
        one_hot = np.zeros(num_classes, dtype=int)
        one_hot[label] = 1

        # Store extracted values
        all_labels.append(one_hot)
        all_pixels.append(pixels)

        # Write one-hot line (integers)
        f_out.write(" ".join(str(x) for x in one_hot) + "\n")
        # Write pixel line (floats)
        f_out.write(" ".join(f"{x:.6f}" for x in pixels) + "\n")

# Convert to numpy arrays
all_labels = np.array(all_labels)  # Shape: (num_samples, 10)
all_pixels = np.array(all_pixels)  # Shape: (num_samples, 784)

# Save labels to separate file
with open(labels_file, "w") as f_labels:
    for label in all_labels:
        f_labels.write(" ".join(str(x) for x in label) + "\n")

# Save pixels to separate file
with open(pixels_file, "w") as f_pixels:
    for pixel in all_pixels:
        f_pixels.write(" ".join(f"{x:.6f}" for x in pixel) + "\n")

print(f"MNIST dataset saved as {output_file}")
print(f"Labels saved as {labels_file}")
print(f"Pixels saved as {pixels_file}")
print(f"\n--- Extraction Summary ---")
print(f"  Samples extracted : {len(all_labels)}")
print(f"  Labels shape      : {all_labels.shape}")
print(f"  Pixels shape      : {all_pixels.shape}")
print(f"\nExample - First sample:")
print(f"  One-hot label: {all_labels[0]}")
print(f"  Pixel values (first 10): {all_pixels[0][:10]}")
```
